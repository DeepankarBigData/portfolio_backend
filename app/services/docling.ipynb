```python
from docling.document_converter import DocumentConverter
import os
from pathlib import Path

# --- Configuration ---
# Set your Hugging Face token as an environment variable, if required by Docling's models
# os.environ["HF_TOKEN"] = "hf_YOUR_TOKEN_HERE" 

source_pdf = "docs/DeepankarResume.pdf"
output_md = "output_document.md"

# --- Conversion Logic ---
converter = DocumentConverter()
result = converter.convert(source_pdf) # The result contains a structured document object

# Export the document to a Markdown string
markdown_content = result.document.export_to_markdown()

# Save the Markdown content to a file
with open(output_md, "w", encoding="utf-8") as f:
    f.write(markdown_content)

print(f"Successfully converted {source_pdf} to {output_md}")
```
